```python
chat_template = ChatPromptTemplate.from_messages([
    ("system", "너는 {language} 언어 선생님 역할을 수행할거야."),
    ("human", "안녕, 오늘 배울 내용은 뭐니?"),
    ("ai", "안녕하세요! 오늘은 기초 문법을 배워볼까요?"),
    ("human", "{user_input}")
])

msg_1 = chat_template.invoke({
    'language': '영어',
    'user_input': '숫자 세는 방법에 대해서 알려줘'
})

for m in msg_1.messages:
    print(m.content)

```

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()

model = ChatGroq(model='llama-3.1-8b-instant')

chat_template = ChatPromptTemplate.from_messages([
    ("system", "너는 {language} 언어 선생님 역할을 수행할거야."),
    ("human", "안녕, 오늘 배울 내용은 뭐니?"),
    ("ai", "안녕하세요! 오늘은 기초 문법을 배워볼까요?"),
    ("human", "{user_input}")
])

result = model.invoke(
    chat_template.invoke({
        'language': '중국어',
        'user_input': '요일 세는 방법에 대해서 알려줘'
    })
)

print(result)
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate.from_template('대전에 대해서 100글자로 소개해줘')

model = ChatGroq(model='llama-3.1-8b-instant')

parser = StrOutputParser()

# prompt -> model -> parser ->

result = parser.invoke(
    model.invoke(
        prompt.invoke({})
    )
)

print(result)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain_core.output_parsers import StrOutputParser

model = ChatGroq(model='llama-3.1-8b-instant')

prompt = ChatPromptTemplate.from_messages([
    ("system", "너는 관광가이드 역할을 수행할거야. 일본어로만 대답해"),
    ("human", "{country}의 {city}에서 가장 유명한 {target}을 3개만 소개해줘")
])

parser = StrOutputParser()

chain = prompt | model | parser

result = chain.invoke({
    'country': '한국',
    'city': '대전',
    'target': '음식'
})

print(result)
```
