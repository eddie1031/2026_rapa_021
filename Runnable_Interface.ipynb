```python
from langchain_core.runnables import RunnableLambda

append_ten = lambda x: x + 10

run_append = RunnableLambda(append_ten)
run_double = RunnableLambda(lambda x: x * 2)

result = run_append.invoke(10)
print(f'result = {result}')

result2 = run_double.invoke(2)
print(f'result2 = {result2}')


result = run_append.invoke(5)
result3 = run_double.invoke(result)
print(f'result3 = {result3}')
```

```python
import asyncio
import time

from langchain_core.runnables import RunnableLambda

import nest_asyncio
nest_asyncio.apply()

async def append_ten(x):
    print(f'{x} append_ten 시작')
    await asyncio.sleep(2)
    print(f'{x} append_ten 완료')
    return x + 10

run_async_append = RunnableLambda(append_ten)

async def main():

    start = time.time()

    results = await asyncio.gather(
        run_async_append.ainvoke(1),
        run_async_append.ainvoke(3),
        run_async_append.ainvoke(5),
        run_async_append.ainvoke(7),
    )

    end = time.time()

    print(f'result = {results}')
    print(f'총 실행시간 = {end - start}')

asyncio.run(main())
```

```python
import time

from langchain_core.runnables import RunnableLambda


def append_ten(x):
    print(f'{x} append_ten 시작')
    time.sleep(2)
    print(f'{x} append_ten 완료')
    return x + 10

run_append = RunnableLambda(append_ten)

start = time.time()

run_append.invoke(1)
run_append.invoke(3)
run_append.invoke(5)
run_append.invoke(7)

end = time.time()

print(f'총 실행시간 = {end - start}')
```

## batch()

```python
from langchain_core.runnables import RunnableLambda

append_ten = lambda x: x + 10

run_append = RunnableLambda(append_ten)

arr = [ _ for _ in range(2, 8, 2) ]

result = run_append.batch(arr)
print(result)
```

## stream()

```python
import time
from langchain_core.runnables import RunnableLambda

def print_str(data: str):
    for _ in data:
        time.sleep(.5)
        yield _

run_print_str = RunnableLambda(print_str)

for _ in run_print_str.stream("안녕하십니까 RAPA 백엔드 여러분 늘 고생 많습니다 힘내세요!"):
    print(_, end="", flush=True)
```
