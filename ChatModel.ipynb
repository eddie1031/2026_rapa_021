## `dotenv`를 사용해서 env 파일로서 환경변수 관리

````python
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("GROQ_API_KEY")
````

````python
from langchain_groq import ChatGroq

# 만약에, 환경변수 이름이 GROQ_API_KEY라면 api_key를 굳이 세팅하지 않아도
# 객체 생성시점에 자동으로 감지하기때문에 문제없이 객체를 사용할 수 있다.
model = ChatGroq(
    model='llama-3.1-8b-instant',
    # api_key=API_KEY,
)

model.invoke('인공지능이 뭐야?')
````

````python
q = [
    '대전의 맛있는 음식 1개를 이름만 알려줘',
    '대전의 유명한 관광지 1곳을 이름만 알려줘'
]

results = model.batch(q)

print(results)
````

````python
query = '대전에서 가장 유명한 빵가게 이름이 뭐야?'

for chunk in model.stream(query):
    print(chunk.content, end='||', flush=True)
````

````python
model = ChatGroq(
    model='llama-3.1-8b-instant',
    temperature=0.1
)

model.invoke('대전이 뭐야?')
````

````python
model = ChatGroq(
    model='llama-3.1-8b-instant',
    temperature=1.0
)

model.invoke('대전이 뭐야?')
````

````python
model = ChatGroq(
    model='llama-3.1-8b-instant',
    max_tokens=100
)

model.invoke('성심당이 뭐야?')
````
